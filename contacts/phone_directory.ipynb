```python
from tkinter import *
from tkinter import ttk
import sqlite3

def contact():
    win=Tk()
    win.title("Contacts list")
    win.geometry('1200x650')
    win.config(bg='#ffff00')
    t=Label(win,text="Phone Directory",font=("times new roman",20,"bold"),fg='red',bg="silver").place(x=500,y=1)
    
    f1=Frame(win,bg='#de5bf0').place(x=5,y=75,height=530,width=380)
    
    img1=PhotoImage(file="name.png")
    c1=Label(f1,height=12,width=12,image=img1)
    c1.place(x=13,y=100)
    name=Label(f1,text="Name",font=("Bahnschrift",18),bg='#de5bf0').place(x=30,y=90)
    
    num=Label(f1,text="Phone",font=("Bahnschrift",18),bg='#de5bf0').place(x=30,y=140)
    img2=PhotoImage(file="phone.png")
    c2=Label(f1,height=12,width=12,image=img2)
    c2.place(x=13,y=150)
    
    email=Label(f1,text="Email",font=("Bahnschrift",18),bg='#de5bf0').place(x=30,y=190)
    img3=PhotoImage(file="email.png")
    c3=Label(f1,height=12,width=12,image=img3)
    c3.place(x=13,y=200)
    
    city=Label(f1,text="City",font=("Bahnschrift",18),bg='#de5bf0').place(x=30,y=240)
    img4=PhotoImage(file="city.png")
    c4=Label(f1,height=15,width=12,image=img4,bg='#de5bf0')
    c4.place(x=13,y=250)
    
    dob=Label(f1,text="Age",font=("Bahnschrift",18),bg='#de5bf0').place(x=30,y=290)
    img5=PhotoImage(file="date.png")
    c4=Label(f1,height=15,width=12,image=img5,bg='pink')
    c4.place(x=13,y=300)
    
    i1=Entry(f1,width=25,bg='white',font=("Bahnschrift",13))
    i1.place(x=120,y=96)
    i2=Entry(f1,width=25,bg='white',font=("Bahnschrift",13))
    i2.place(x=120,y=146)
    i3=Entry(f1,width=25,bg='white',font=("Bahnschrift",13))
    i3.place(x=120,y=196)
    i4=Entry(f1,width=25,bg='white',font=("Bahnschrift",13))
    i4.place(x=120,y=246)
    i5=Entry(f1,width=25,bg='white',font=("Bahnschrift",13))
    i5.place(x=120,y=296)
    
    note=Label(win,text="*** note : Name and E-mail cannot be edited ***",font=("Bahnschrift",14),fg="red")
    note.place(x=390,y=618)
    
    def cleartext():
        i1.config(state=NORMAL)
        i3.config(state=NORMAL)
        i1.delete(0,END)
        i2.delete(0,END)
        i3.delete(0,END)
        i4.delete(0,END)
        i5.delete(0,END)
        
    def insertdata():
        con=sqlite3.connect('new_phonedirectory.db',timeout=10)
        cur=con.cursor()
        cur.execute('CREATE TABLE IF NOT EXISTS DIRECTORY(name text,number text,email text,city text,age text)')
        cur.execute('INSERT INTO DIRECTORY(name,number,email,city,age) VALUES(?,?,?,?,?)',(i1.get(),i2.get(),i3.get(),i4.get(),i5.get()))
        con.commit()
        con.close()
        cleartext()
        display()
        
        
    def display():
        databox.delete(0,END)
        con=sqlite3.connect('new_phonedirectory.db',timeout=10)
        cur=con.cursor()
        query='SELECT * FROM DIRECTORY'
        cur.execute(query)
        rows=cur.fetchall()
        for i in rows:
            databox.insert(END,i)
        con.commit()
        con.close()
        
    def selection_fun(event=NONE):
        global pd
        cleartext()
        searchpd=databox.curselection()[0]
        pd=databox.get(searchpd)
        i1.insert(END,pd[0])
        i2.insert(END,pd[1])
        i3.insert(END,pd[2])
        i4.insert(END,pd[3])
        i5.insert(END,pd[4])
        i1.config(state=DISABLED)
        i3.config(state=DISABLED)
        
    def deletefun():
        con=sqlite3.connect('new_phonedirectory.db',timeout=10)
        cur=con.cursor()
        arg1=i1.get()
        arg2=i2.get()
        cur.execute('DELETE FROM DIRECTORY WHERE name=? and number=?',(arg1,arg2))
        con.commit()
        con.close()
        cleartext()
        display()
    
    def updatefun():
        con=sqlite3.connect('new_phonedirectory.db',timeout=10)
        cur=con.cursor()
        arg1=i1.get()
        arg2=i2.get()
        arg3=i3.get()
        arg4=i4.get()
        arg5=i5.get()
        cur.execute('UPDATE DIRECTORY SET number=?,city=?,age=? WHERE name=? and email=?',(arg2,arg4,arg5,arg1,arg3))
        con.commit()
        con.close()
        cleartext()
        display()
    
    def searchdata():
        databox.delete(0,END)
        con=sqlite3.connect('new_phonedirectory.db',timeout=10)
        cur=con.cursor()
        arg=sentry.get()
        print(tkvar.get()+" "+arg)
        if tkvar.get()== 'name' :
            cur.execute('SELECT * FROM DIRECTORY WHERE name=?',(arg,))
        else:
            cur.execute('SELECT * FROM DIRECTORY WHERE number=?',(arg,))
        rowsearch=cur.fetchall()
        for i in rowsearch:
            databox.insert(END,i)
            print(i)
        con.commit()
        con.close()
        
        
    f2=Frame(f1,bg='#34e8eb').place(x=40,y=350,height=220,width=300)
    
    addimg=PhotoImage(file="add.png")
    upimg=PhotoImage(file="update.png")
    delimg=PhotoImage(file="del.png")
    cleimg=PhotoImage(file="clear.png")
    searchimg=PhotoImage(file="search.png")
    showallimg=PhotoImage(file="show.png")
                     
    add=Button(f2,text="Add",image=addimg,width=80,compound=LEFT,font=("Bahnschrift",12,"bold"),bg="orange",command=insertdata)
    add.place(x=140,y=363)
    updateb=Button(f2,text="Update",image=upimg,width=90,compound=LEFT,font=("Bahnschrift",12,"bold"),bg="orange",command=updatefun)
    updateb.place(x=135,y=408)
    delb=Button(f2,text="Delete",image=delimg,width=90,compound=LEFT,font=("Bahnschrift",12,"bold"),bg="orange",command=deletefun)
    delb.place(x=135,y=453)
    clearb=Button(f2,text="Clear",image=cleimg,width=90,compound=LEFT,font=("Bahnschrift",12,"bold"),bg="orange",command=cleartext)
    clearb.place(x=135,y=503)
    
    dataF=Frame(win,bg="#4dff4d").place(x=420,y=75,height=530,width=752)
    
    search=Label(dataF,text="Search by",font=("Bahnschrift",15,"bold"),bg="#4dff4d").place(x=425,y=80)
    tkvar = StringVar(dataF)
    choices = {'name','number'}
    tkvar.set('name')
    box=OptionMenu(dataF,tkvar,*choices)
    box.config(bg='#ff4d4d',font=("Bahnschrift",12))
    box.place(x=540,y=80)
    sentry=Entry(dataF,width=20,font=("Bahnschrift",12))
    sentry.place(x=650,y=85)
    bsearch=Button(dataF,text="Search",image=searchimg,font=("Bahnschrift",10,"bold"),compound=LEFT,command=searchdata,bg="orange")
    bsearch.place(x=870,y=80)
    sall=Button(dataF,text="Show All",image=showallimg,font=("Bahnschrift",10,"bold"),compound=LEFT,bg="orange",command=display)
    sall.place(x=970,y=80)
    
    scrollx=Scrollbar(win,orient=HORIZONTAL)
    scrolly=Scrollbar(win,orient=VERTICAL)
    databox=Listbox(win,bg='white',xscrollcommand=scrollx.set,yscrollcommand=scrolly.set,font=("Bahnschrift",20))
    databox.bind('<<ListboxSelect>>',selection_fun)
    databox.place(x=423,y=128,height=457,width=728)
    
    scrollx.pack(side=BOTTOM,fill=X)
    scrolly.pack(side=RIGHT,fill=Y)
    scrollx.place(x=423,y=585,width=744)
    scrolly.place(x=1150,y=128,height=460)
    scrollx.config(command=databox.xview)
    scrolly.config(command=databox.yview)
    win.mainloop()
    
    
contact()
        
```
